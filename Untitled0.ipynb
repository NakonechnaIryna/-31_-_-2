```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Генерація випадкового набору даних в діапазоні 0-1000
random_data = np.random.randint(0, 1000, size=1000)

# Нормалізація даних до діапазону 0-1
normalized_data = (random_data - np.min(random_data)) / (np.max(random_data) - np.min(random_data))

# Розділення на навчальну і тестову вибірки
X_train, X_test, y_train, y_test = train_test_split(normalized_data, normalized_data, test_size=0.2, random_state=42)

best_k = None
best_mse = float('inf')
mse_values = []

# Навчання KNN-регресора з різними значеннями K та вибір найкращого K
for k in range(1, 21):
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(X_train.reshape(-1, 1), y_train)
    y_pred = knn.predict(X_test.reshape(-1, 1))
    mse = mean_squared_error(y_test, y_pred)
    mse_values.append(mse)
    if mse < best_mse:
        best_mse = mse
        best_k = k

# Візуалізація результатів
plt.plot(range(1, 21), mse_values, marker='o')
plt.xlabel('K')
plt.ylabel('Середньоквадратична помилка (MSE)')
plt.title('Залежність MSE від K для KNN Regressor')
plt.grid(True)
plt.show()

print(f"Найкраще значення K: {best_k}, Найкращий MSE: {best_mse}")
```
